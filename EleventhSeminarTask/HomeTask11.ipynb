```python
from sympy import *
from sympy.plotting import plot
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = Symbol('x')
 #Заданная функция №1
func1 = x**3 - 50*x                  
#Заданная функция №2
func2 = -x**4 + 88*x**2 - 241        
#Функция получаемая сложением выражений
func3 = func1 + func2                
#Функция полученная в результате приравнивания №1 и №2 для нахождения пересечений
func4 = -x**4 + 88*x**2 - 241 - x**3 + 50*x 
```

```python
#Находим точки пересечения заданных функций func1 и func2
```

```python
x1,x2,x3,x4 = solve(func4)
```

```python
x1 = x1.evalf(chop = True)
x1
```

```python
x2 = x2.evalf(chop = True)
x2
```

```python
x3 = x3.evalf(chop = True)
x3
```

```python
x4 = x4.evalf(chop = True)
x4
```

```python
y1 = x1**3 - 50*x1
y1 = y1.evalf(chop = True)
y1
```

```python
y2 = x2**3 - 50*x2
y2 = y2.evalf(chop = True)
y2
```

```python
y3 = x3**3 - 50*x3
y3 = y3.evalf(chop = True)
y3
```

```python
y4 = x4**3 - 50*x4
y4 = y4.evalf(chop = True)
y4
```

```python
#Добавляем координаты точек пересечения в списки
```

```python
coord_x = [x1,x2,x3,x4]
coord_y = [y1,y2,y3,y4]
```

```python
#Строим графики заданных функций
```

```python
function1 = x**3 - 50*x
function2 = -x**4 + 88*x**2 - 241
function3  = function1 + function2
x = np.linspace(-10, 10, 1000)
fig = plt.figure(figsize = (20, 5))

plt.plot(x, function1,'r',label = 'x**3 - 50*x')
plt.plot(x, function2,'k',label = '-x**4 + 88*x**2 - 241')
plt.plot(x, function3,'--',label = '-x**4 - x**3 + 88*x**2 + 50*x -241')
plt.plot(coord_x, coord_y, 'o')
plt.grid()
plt.legend(fontsize = 20)


plt.show()
```
